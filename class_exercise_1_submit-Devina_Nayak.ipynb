### Exercise 1.15: find the start time of Einstein’s presentation
Suppose a presentation session at a conference starts at 8:25am, each speaker is given 25 minutes and they present in the ascending order of their surname, when does Einstein start to present?
- There will be no duplicate surnames
- Take the last word as the surname

````python
physicists = ['Erwin Schroedinger', 'Wolfgang Pauli', 'Max Born', 'Niels Bohr', 'Max Planck', 'Madame Curie', 'Hendrik Antoon Lorentz', 'Albert Einstein', 'Paul Langevin', 'Louis Victor de Broglie', 'Paul Dirac', 'Werner Heisenberg']
````

````python
surname_list=[]
for item in physicists:
    surname_list.append(item.split(" ")[-1])

sorted_surnames = sorted(surname_list)
index = sorted_surnames.index('Einstein')
elapsed_minute = 25*index
hour_component = 25*index / 60
remainder_component = elapsed_minute % 60
final_hour = str(int(8 + hour_component))
final_min = str(int(25 + remainder_component))
print(final_hour+":"+final_min)
````
